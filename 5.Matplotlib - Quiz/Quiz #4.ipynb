## Quiz4.สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - - 

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,30)
y = np.cos(x)
plt.plot( x , y , c= 'R' , ls = '--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = COS(X)')
```
