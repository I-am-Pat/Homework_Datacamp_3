## Quiz3.สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.arange(0,30)
x = y**2
plt.plot( x ,y , c= '#03c4a1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X = Y^2')
```
