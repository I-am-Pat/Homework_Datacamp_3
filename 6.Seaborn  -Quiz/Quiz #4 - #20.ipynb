##### ให้ import ไฟล์ titanic ไฟล์เดิมจากบทเรียน Pandas เข้าสู่

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\SUPAT\anaconda3\Data Science Learning\Dataset\Titanic - Train Data.csv')
df = df.dropna()
df.head()
df['Age'] = df['Age'].astype(int)
df
```

## 4. สร้าง count plot นับข้อมูล Embarked โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
df['Embarked'].replace({'C':'Cherbourg','Q':'Queenstown','S':'Southampton'} , inplace = True)
```

```python
# Vertical
fig = plt.figure(figsize = [10,8])
ax = sns.countplot(x = 'Embarked' , data = df)
plt.ylim(0,140)
x = np.flip(df['Embarked'].value_counts(sort=False).values)
    
for index , i in enumerate(x) :
    ax.annotate(i, xy=(index, i+3), size = 15, verticalalignment='center_baseline')
```

```python
# Horizontal
fig = plt.figure(figsize = [10,8])
ax = sns.countplot(y = 'Embarked' , data = df)
plt.xlim(0,140)


for index , i in enumerate(x) :
    ax.annotate(i, xy=(i+3,index), size = 15, verticalalignment='center_baseline')
```

## 5. สร้าง count plot นับข้อมูล Pclass โดยสร้างกราฟแบบแนวตั้งและแนวนอน

```python
df['Pclass'].value_counts()
df['Pclass'].replace({1:'1st',2:'2nd',3:'3rd'},inplace = True)
```

```python
# Vertical
fig = plt.figure(figsize = [10,8])
order = ['1st','2nd','3rd']
ax = sns.countplot(x = 'Pclass',data = df , order = order)
plt.ylim(0,180)

x = df['Pclass'].value_counts()

for index , i in enumerate(order) :
    ax.annotate(x[i],xy = (index , x[i]+5) , size = 15 , verticalalignment = 'center_baseline' , horizontalalignment = 'center')
```

```python
# Horizontal
fig = plt.figure(figsize = [10,8])
order = ['1st','2nd','3rd']
ax = sns.countplot(y = 'Pclass',data = df , order = order)
plt.xlim(0,180)

x = df['Pclass'].value_counts()

for index , i in enumerate(order) :
    ax.annotate(x[i],xy = (x[i]+5,index) , size = 15 , verticalalignment = 'center_baseline' , horizontalalignment = 'center')
```

## 11. สร้าง Stripplot โดยให้ แกน X เป็น Survived แกน Y เป็นอายุ และ hue เป็นเพศ

## 13. สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ 

```python
Result = pd.crosstab(df['Pclass'],df['Sex'])
fig = plt.figure(figsize = [12,8])
sns.heatmap(Result , cmap = 'coolwarm' , annot = True , annot_kws ={'size':15})
```

## 14. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา# 

```python
Result = df.pivot_table(index = 'Pclass' , columns = 'Sex' , values = 'Fare' , aggfunc = np.median)
fig = plt.figure(figsize = [12,8])
sns.heatmap(Result , cmap = 'twilight_shifted' , annot = False , linecolor = 'white' , linewidth = 0.25)
```

## 15. สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง

```python
Result = df.pivot_table(index = 'Pclass' , columns = 'Sex' , values = 'Fare' , aggfunc = np.mean)
fig = plt.figure(figsize = [12,8])
sns.heatmap(Result , cmap = 'coolwarm' , annot = True , annot_kws = {'size':12} , linecolor = 'white' , linewidth = 2)
```
